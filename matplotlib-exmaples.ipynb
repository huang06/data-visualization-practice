```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("seaborn")
```

## Subplots

### Create common layouts of subplots

```python
fig, axs = plt.subplots(2, 2)
```

### Create axes to arbitrary location of the figure

> The dimensions [left, bottom, width, height] of the new Axes. All quantities are in fractions of figure width and height.

```python
fig = plt.figure(figsize=(14, 7))
ax1 = fig.add_axes((0.0, 0.0, 1.0, 1.0), xticks=np.arange(0.0, 1.1, 0.1), yticks=np.arange(0.0, 1.1, 0.1))
ax2 = fig.add_axes(
    (0.2, 0.2, 0.6, 0.6), facecolor="coral", xticks=np.arange(0.0, 1.1, 0.1), yticks=np.arange(0.0, 1.1, 0.2)
)
ax3 = fig.add_axes(
    (0.4, 0.4, 0.2, 0.2),
    facecolor="royalblue",
    xticks=np.arange(0.0, 1.1, 0.2),
    yticks=np.arange(0.0, 1.1, 0.2),
)
```
